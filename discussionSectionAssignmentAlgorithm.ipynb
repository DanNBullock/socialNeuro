In this notebook I describe a method for assigning students to tasks/groups in accordance with several criteria.  A brief description of the challenge is as follows:

- There are 36 students in the class
- A total of 12 primary literature articles will be read (which will be covered on distinct days throughout the semester)
- Each student must lead discussion for **two** of the classes and must write a short article summary for **seven** of the classes
- **Ideally**, any day for which a student has written an article summary, that student should not lead a discussion
- Students should **only** lead two discussion sections for the semester
- On discussion days, students are broken into **six** groups of **six** with one discussion leader per group
- Students express their preference by listing the seven papers that they would like to write their article summaries about.
- Because there are not restrictions on this process, it is conceivable that, on any given discussion day, more than **thirty** students would sign up to write about the article, and so you would end up with a shortage of students who are *not writing* and thus would serve as discussion leaders.  In such cases, manual ascription of discussion leaders, in violation of the no writing and discussion leading for the same class rule of thumb, would be necessary.
- It is recomended to order the assigmnent process in such a way that discussion days with the fewest degrees of freedom are handled first (as this prevents dead-end assignment mappings, where not all papers have discussion leaders)

We'll begin this example algorithm by loading an anonymized version of a csv listing the student's preferences

```python
import pandas as pd

#load csv
preferenceFrame=pd.read_csv('anonPreferences.csv',header=None)

#get the shape of the dataframe
preferenceFrameShape=preferenceFrame.shape

#print it out just to look at it
preferenceFrame.head(len(preferenceFrame))
```

In the origional discussion thread, there were a number of variations and spellings in how preferences were communicated.  The following section was used to pare down the unique tokens to the point where there were as many tokens as there were papers (12)

```python
#get rid of the names so you can have a preferences only dataframe
onlyPreferencesFrame=preferenceFrame.drop(labels=0,axis=1)

#get the unique entries, just to check for mispellings
#there should be a total of 12
uniqueEntries=pd.unique(onlyPreferencesFrame.values.flatten())
#print to view
print(uniqueEntries)
```

To get a sense of the frequency of choices, and if there are any papers that will prove to be a challenge, we can make a count plot of preferences.

```python
import seaborn as sns
#chagne the column names
authorSeries=pd.DataFrame(onlyPreferencesFrame.values.flatten(),columns=['Authors'])
#plot it so we can visualize the frequency of choices.
sns.countplot(y="Authors",data=authorSeries)
```

From the plot we can see that Sherman violates the soft limit on 30 students, and so a student will need to be assigned to both lead discussion and write on that day.  This also requires some additional hard coding as will be seen later in this notebook.

Next we can get the explicit counts associated with this plot.

```python
#get the frequency count
authorCounts=authorSeries.iloc[:,0].value_counts()
#reset the indexes
authorCounts=authorCounts.reset_index()
#rename the columns
authorCounts=authorCounts.rename(columns={'index':'authorName','Authors':'count'})
#print to output
authorCounts
```

Just for the sake of viewing the entries, we can print out the students' names.  Because this has been anonymized, this will not be particularly informative.

```python
#get the student's names
allStudentsSeries=preferenceFrame.iloc[:,0]
#print to visualize
print(allStudentsSeries)
```

Our first step will be to iterate across authors/days to determine which students have expressed preferences for authors on which days.  In essence, we will be creating a list of lists, in which the i-th entry of the outer list contains a list of all of the students who have expressed a desire to write a article summary for that author.  Furthermore, the authors in this list will be sorted in accordance with the count structure depicted earlier.  The excepton to this is that we will be dropping the record for Sherman, due to the issue noted earlier (need for manual assignment)

```python
#because we will manually assign sherman
authorCountsNoSherman=authorCounts.drop(labels=0,axis=0)

#establish storage objects
dailyCandidates = []

#iterate across papers
for iAuthors in range(len(authorCountsNoSherman)):
    currentCandidates=[]
    #set the current author
    currentAuthor=authorCountsNoSherman['authorName'].iloc[iAuthors]
    #iterate across students
    for index,value in allStudentsSeries.items():
        #get current student's name              
        currentStudentName=preferenceFrame.loc[preferenceFrame.iloc[:,0]==value,0].item()
        #get the current student's preferences            
        currentPreferences=onlyPreferencesFrame.loc[index,:]
        #determine if current author is in the current student's preferences
        presentBool=onlyPreferencesFrame.loc[index,:].isin([currentAuthor]).any()
        #if the student listed it
        if presentBool:
            #append the current students name to "todays" list
            currentCandidates.append(currentStudentName)
    dailyCandidates.append(currentCandidates)

#view the output, its a mess, but its a good check
print (dailyCandidates)
```

```python
#import these packages as they will be necessary
import numpy as np
import random

#initialize a dataframe for the assignments
assignmentDataFrame=pd.DataFrame(allStudentsSeries)
#rename the column
assignmentDataFrame=assignmentDataFrame.rename(columns={0:'StudentName'})
#create columns for the 
assignmentDataFrame['leadDay1']=''
assignmentDataFrame['leadDay2']=''

#go ahead and do manual assignment for sherman
assignmentDataFrame['leadDay1'].loc[assignmentDataFrame['StudentName']=='StudentName29']='Sherman'
assignmentDataFrame['leadDay1'].loc[assignmentDataFrame['StudentName']=='StudentName20']='Sherman'
assignmentDataFrame['leadDay1'].loc[assignmentDataFrame['StudentName']=='StudentName35']='Sherman'
assignmentDataFrame['leadDay1'].loc[assignmentDataFrame['StudentName']=='StudentName13']='Sherman'
assignmentDataFrame['leadDay1'].loc[assignmentDataFrame['StudentName']=='StudentName36']='Sherman'
assignmentDataFrame['leadDay1'].loc[assignmentDataFrame['StudentName']=='StudentName6']='Sherman'

#iterate across the authors to perform the rest of the assignments
for iAuthors in range(len(authorCountsNoSherman)):
    #get current author name
    currentAuthor=authorCountsNoSherman['authorName'].iloc[iAuthors]
    #get students who aren't writing today
    notWritingToday=list(set(allStudentsSeries)-set(dailyCandidates[iAuthors]))
    #compute those that are available for assignemnt, i.e. haven't gone twice already
    avaibaleBool=np.logical_or(assignmentDataFrame['leadDay1']=='',assignmentDataFrame['leadDay2']=='')
    #use that info to get the actual names
    avaibaleNames=allStudentsSeries.loc[avaibaleBool]
    #find the names of avaiable AND not writing today students
    canBeAssigned=list(set.intersection(set(notWritingToday),set(avaibaleNames)))
    #shuffle the ones that can be assigned
    random.shuffle(canBeAssigned)
    #pick the first six
    toBeAssigned=canBeAssigned[0:6]
    #iterate across the 6 stuents to be assigned
    for itoBeAssigned in range(len(toBeAssigned)):
        #get name of current student to assign
        currentStudentToAssign=toBeAssigned[itoBeAssigned]
        #determine if they are available for day 1
        #because we checked to ensure they hadn't done two days with avaibaleBool, if they fail this, it means they can go on day2
        day1AvaiableBool=assignmentDataFrame['leadDay1'].loc[assignmentDataFrame['StudentName']==currentStudentToAssign]==''
        #
        if day1AvaiableBool.values:
            assignmentDataFrame['leadDay1'].loc[assignmentDataFrame['StudentName']==currentStudentToAssign]=currentAuthor
        #because we did the availability check, this is a safe assumption
        else:
            assignmentDataFrame['leadDay2'].loc[assignmentDataFrame['StudentName']==currentStudentToAssign]=currentAuthor

            
#print the resultant dataframe with assignments
print(assignmentDataFrame)
```
